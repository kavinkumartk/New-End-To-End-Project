```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 

df=pd.read_csv('train.csv')
df
```

```python
df.shape
```

```python
df1=df.sample(n=3000, random_state=42)  
print(df1.shape)
```

```python
df1.isnull().sum()
```

```python
df1.fillna(df['Arrival Delay in Minutes'].mean(),inplace=True)
df1.isnull().sum()
```

```python
df1.duplicated().sum()
```

```python
df.info()
```

```python
df1.describe()
```

```python
df1=df1.drop(columns=['Unnamed: 0', 'id'])
df1
```

```python
from sklearn.preprocessing import LabelEncoder
label_cols = ['Gender', 'Customer Type', 'Type of Travel', 'Class','satisfaction']
label_encoder = LabelEncoder()

for col in label_cols:
    df1[col] = label_encoder.fit_transform(df1[col])
df1
```

```python
Q1 = df1.quantile(0.25)
Q3 = df1.quantile(0.75)
IQR = Q3 - Q1

outliers = ((df1 < (Q1 - 1.5 * IQR)) | (df1 > (Q3 + 1.5 * IQR)))
print("Outliers count per column:\n", outliers.sum())
```

```python
df1.skew()
```

```python
df1.hist(bins=30, figsize=(15, 10), grid=False)
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import PowerTransformer, RobustScaler
import pandas as pd

skewed_cols = ['Departure Delay in Minutes', 'Arrival Delay in Minutes', 'Flight Distance']

pt = PowerTransformer(method='yeo-johnson')
df1[skewed_cols] = pt.fit_transform(df1[skewed_cols])

outlier_cols=['Arrival Delay in Minutes','Departure Delay in Minutes','Inflight service','Flight Distance']

scaler = RobustScaler()
df1[outlier_cols] = scaler.fit_transform(df1[outlier_cols])
```

```python
df1.hist(bins=30, figsize=(15, 10), grid=False)
plt.tight_layout()
plt.show()
```
